```python
import matplotlib.pyplot as plt
import pandas as pd
import requests
```

```python
data2 = pd.read_excel(r"C:\Users\nasta\OneDrive\Документы\hw-03-MarchenkovaAnastasya\first_stat.xlsx")
data2
data1 = pd.read_excel(r"C:\Users\nasta\OneDrive\Документы\hw-03-MarchenkovaAnastasya\second_stat.xlsx")
data1
```

```python
data1 = data1.rename(columns={'Таблица аналитическая': 'Округ'})
data2 = data2.rename(columns={'Таблица аналитическая': 'Округ'})
data1
```

Решейпинг массива


```python
data_old = pd.merge(data1, data2, on='Округ')

data = data_old.rename(columns={'Родившиеся в 2023 (на тыс. населения)': 2023, 'Родившиеся в 2022 (на тыс. населения)': 2022}) #переименовываю, чтобы при мелте в столбце год был только год, без "Родившиеся на тыс.человек"
data
```

```python
data_quasilong = data.melt(id_vars = data.columns[0],
                         value_vars = data.columns[7:9],

                         var_name = 'Год',
                         value_name = 'Количество родившихся (на тыс. населения)')
data_quasilong
```

Проанализируем рождаемость за 2023 год по субъектам


```python
federal_districts = [
    "Российская Федерация",
    "Центральный федеральный округ",
    "Северо-Западный федеральный округ",
    "Южный федеральный округ",
    "Северо-Кавказский федеральный округ",
    "Приволжский федеральный округ",
    "Уральский федеральный округ",
    "Сибирский федеральный округ",
    "Дальневосточный федеральный округ"
]
data_quasilong['Округ'] = data_quasilong['Округ'].astype(str)
data_quasilong['Год'] = data_quasilong['Год'].astype(int)
data_quasilong['Округ'] = data_quasilong['Округ'].str.strip()

data2023 = data_quasilong.loc[(data_quasilong['Год'] == 2023) & ~(data_quasilong['Округ'].isin(federal_districts))]
data2023
```

```python
data2023["Количество родившихся (на тыс. населения)"].describe()
```

```python
data2023["Количество родившихся (на тыс. населения)"].median()
```

Построим ящик с усами

```python
plt.boxplot(data2023["Количество родившихся (на тыс. населения)"])
```

Наблюдаются нехарактерные значения, посмотрим в каких субъектах

```python
delt = data2023["Количество родившихся (на тыс. населения)"].describe()[6] - data2023["Количество родившихся (на тыс. населения)"].describe()[4]
```

```python
xvgh = data2023["Количество родившихся (на тыс. населения)"].describe()[6] + 1.5 * delt #верхняя граница характерных значений
xvgh
```

Регионы со значениями выше верхнего характерного значения:

```python
res = data2023.loc[data2023["Количество родившихся (на тыс. населения)"] > xvgh ]
res
```

Построим гистограмму

```python
plt.xlabel('Количество родившихся (на тыс. населения)')
plt.ylabel('Количество значений')
plt.hist(data2023["Количество родившихся (на тыс. населения)"], color = 'blue', edgecolor = 'black',
         bins = int(5))
```
